# Data Cleaning

```python
import pandas as pd
import re
```

```python
df = pd.read_excel(r"Customer Call List.xlsx")
```

```python
df
```

```python
df=df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
df
```

```python
df["Last_Name"] = df["Last_Name"].str.replace(r'[._/]+', '')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python
df['Phone_Number'] = df['Phone_Number'].replace(['nan--', 'Na--'], '')
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace({'Y': 'Yes', 'N': 'No', 'N/a': ''})
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y': 'Yes', 'N': 'No'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('No')
df
```

```python
df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
df = df[df['Do_Not_Contact'] != 'Yes']
df
```

```python
df = df[df['Phone_Number'] != '']
df
```

```python
df = df.reset_index(drop = True)
df
```

# Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr(), annot = True)
plt.rcParams['figure.figsize'] = (30,7)
plt.show()
```

```python
df3 = df2.groupby('Continent')[df2.columns[5:13]].mean().sort_values(by='2022 Population', ascending = False)
df3
```

```python
ordered_columns = sorted(df3.columns)
df3 = df3.reindex(columns=ordered_columns)
df3
```

```python
df2[df2['Continent'].str.contains('Oceania')]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
df3.plot(kind='bar', ax=ax)
ax.set_title('Population Trend by Continent')
ax.set_xlabel('Continent')
ax.set_ylabel('Population')
ax.grid(True)
ax.legend(title='Year')
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

df3.T.plot(kind='line', ax=ax, linewidth=2)

ax.set_title('Population Trend by Continent')
ax.set_xlabel('Year')
ax.set_ylabel('Population')
ax.grid(True)
ax.legend(title='Continent')
plt.xticks(rotation=45) 
ax.grid(False) 
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))


df3.boxplot(ax=ax)

ax.set_title('Population Distribution by Continent')
ax.set_ylabel('Population')
ax.set_xlabel('Continent')

plt.xticks(rotation=45) 

plt.tight_layout()
plt.show()
```
